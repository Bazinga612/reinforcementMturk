```python
import os
import sys
import csv
from sqlalchemy import Column, ForeignKey, Integer, String
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import relationship
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
from sqlalchemy.sql import select
 
Base = declarative_base()
 
class Dialog_acts(Base):
    __tablename__ = 'dialog_acts'
    id = Column(Integer, primary_key=True)
    dialog_acts = Column(String(250), nullable=False)

class Starter(Base):
    __tablename__ = 'starter'
    id = Column(Integer, primary_key=True)
    utterance = Column(String(250), nullable=False)
 
 
class Dialog_act1(Base):
    __tablename__ = 'dialog_act1'
    id = Column(Integer, primary_key=True)
    utterance = Column(String(250), nullable=False)
    name_dialog_act1 = Column(Integer, ForeignKey('dialog_acts.dialog_acts'))
    dialog_acts = relationship(Dialog_acts)

class Dialog_act2(Base):
    __tablename__ = 'dialog_act2'
    id = Column(Integer, primary_key=True)
    utterance = Column(String(250), nullable=False)
    name_dialog_act2 = Column(Integer, ForeignKey('dialog_acts.id'))
    dialog_acts = relationship(Dialog_acts)

class Dialog_act3(Base):
    __tablename__ = 'dialog_act3'
    id = Column(Integer, primary_key=True)
    utterance = Column(String(250), nullable=False)
    name_dialog_act3 = Column(Integer, ForeignKey('dialog_acts.id'))
    dialog_acts = relationship(Dialog_acts)

class Dialog_act4(Base):
    __tablename__ = 'dialog_act4'
    id = Column(Integer, primary_key=True)
    utterance = Column(String(250), nullable=False)
    name_dialog_act4 = Column(Integer, ForeignKey('dialog_acts.id'))
    dialog_acts = relationship(Dialog_acts)
    
class Mappings(Base):
    __tablename__ = 'mappings'
    mapping_id = Column(Integer, primary_key=True)
    starter_id = Column(Integer, ForeignKey('starter.id'))
    starter = relationship(Starter)
    dialog_act1_id = Column(Integer, ForeignKey('dialog_act1.id'),nullable=True)
    dialog_act1 = relationship(Dialog_act1)
    dialog_act2_id = Column(Integer, ForeignKey('dialog_act2.id'),nullable=True)
    dialog_act2 = relationship(Dialog_act2)
    dialog_act3_id = Column(Integer, ForeignKey('dialog_act3.id'),nullable=True)
    dialog_act3 = relationship(Dialog_act3)
    dialog_act4_id = Column(Integer, ForeignKey('dialog_act4.id'),nullable=True)
    dialog_act4 = relationship(Dialog_act4)

engine = create_engine('sqlite:///demo100.db')

Base.metadata.create_all(engine)

Base.metadata.bind = engine
 
DBSession = sessionmaker(bind=engine)

session = DBSession()

#inserting values in dialog_acts table
list_dialog_acts=['Answer.negative','Answer.positive','Answer.rhetorical','Answer.TalkMore','Answer.change','Answer.continuation','Answer.elaboration','Answer.joking_sarcastic']
for value in list_dialog_acts:
    new_d1 = Dialog_acts(dialog_acts=value)
    session.add(new_d1)
    session.commit()

conn = engine.connect()
with open('tv_movie_combined.csv') as csvDataFile:
    csvReader = csv.reader(csvDataFile)
    next(csvReader, None)
    for row in csvReader:
        #populating Starter table
        select1=select([Starter.utterance]).where(Starter.utterance ==row[4])
        result1=conn.execute(select1)
        fetchone1 = result1.fetchone()
        
        if fetchone1 is None:
            
            new_d = Starter(utterance=row[4])
            session.add(new_d)
            session.commit()
            select1=''
        #populating Dialog_acts1 table

        select2=select([Dialog_act1.utterance]).where(Dialog_act1.utterance ==row[5])
        result2=conn.execute(select2)
        fetchone2 = result2.fetchone()
        
        if fetchone2 is None:
            
            new_da = Dialog_act1(utterance=row[5],name_dialog_act1=row[0])
            session.add(new_da)
            session.commit()
            select2=''
        
        #populating Dialog_acts2 table

        select3=select([Dialog_act2.utterance]).where(Dialog_act2.utterance ==row[6])
        result3=conn.execute(select3)
        fetchone3 = result3.fetchone()
        
        if fetchone3 is None:
            
            new_da = Dialog_act2(utterance=row[6],name_dialog_act2=row[1])
            session.add(new_da)
            session.commit()
            select3=''
 
        #populating Dialog_acts3 table

        select4=select([Dialog_act3.utterance]).where(Dialog_act3.utterance ==row[7])
        result4=conn.execute(select4)
        fetchone4 = result4.fetchone()
        
        if fetchone4 is None:
            
            new_da = Dialog_act3(utterance=row[7],name_dialog_act3=row[2])
            session.add(new_da)
            session.commit()
            select4=''
        
        #populating Dialog_acts4 table

        select5=select([Dialog_act4.utterance]).where(Dialog_act4.utterance ==row[8])
        result5=conn.execute(select5)
        fetchone5 = result5.fetchone()
        
        if fetchone5 is None:
            
            new_da = Dialog_act4(utterance=row[8],name_dialog_act4=row[3])
            session.add(new_da)
            session.commit()
            select5=''
        
        #populating Mappings table
        mappings_table()
        
```

```python
def mappings_table():
    select2=select([Dialog_act1.id]).where(Dialog_act1.utterance ==row[5])
    select3=select([Dialog_act2.id]).where(Dialog_act2.utterance ==row[6])
    select4=select([Dialog_act3.id]).where(Dialog_act3.utterance ==row[7])
    select5=select([Dialog_act4.id]).where(Dialog_act4.utterance ==row[8])
    select1=select([Starter.id]).where(Starter.utterance ==row[4])
    new_mapping = Mappings(starter_id=conn.execute(select1).fetchone()[0],dialog_act1_id=conn.execute(select2).fetchone()[0],dialog_act2_id=conn.execute(select3).fetchone()[0],dialog_act3_id=conn.execute(select4).fetchone()[0]
                          ,dialog_act4_id=conn.execute(select5).fetchone()[0])
    session.add(new_mapping)
    session.commit()
    
```
